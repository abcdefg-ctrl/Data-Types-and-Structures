#Assignment 1: Data Types and Structures


1.	What are data structures, and why are they important?

   -> Data structures are specialized formats for organizing, processing, and storing data. They are crucial because they help manage large amounts of data efficiently, making it easy to access, modify, and perform operations on the data.
2.	Explain the difference between mutable and immutable data types with examples.

   ->	Mutable data types can be changed after creation, like lists and dictionaries. Immutable types, such as tuples and strings, cannot be altered once created. For example, a list can be updated with new elements, but a tuple cannot.
3.	What are the main differences between lists and tuples in Python?

 ->	Lists are mutable, meaning they can be modified after creation. Tuples are immutable. Lists use more memory but are more flexible, whereas tuples are more memory-efficient and used for fixed collections.
4.Describe how dictionaries store data.

 -> Dictionaries store data in key-value pairs. Each key maps to a specific value, and keys must be unique and immutable.
5.	Why might you use a set instead of a list in Python?

  -> Sets do not allow duplicate values, making them ideal for membership testing and eliminating duplicates. They also provide faster lookup times than lists.
6.	What is a string in Python, and how is it different from a list?

 -> A string is a sequence of characters, immutable in nature. A list is mutable and can contain items of different data types.
7.	How do tuples ensure data integrity in Python?

 -> Since tuples are immutable, their content cannot be changed, which makes them useful for ensuring the integrity of data that should not be altered.
8.	What is a hash table, and how does it relate to dictionaries in Python?

 -> A hash table stores data using a hash function that maps keys to values. Python's dictionaries use hash tables to provide efficient key-based access.
9.	Can lists contain different data types in Python?

 -> Yes, Python lists can contain elements of various data types, such as integers, strings, and even other lists or objects.
10.	Explain why strings are immutable in Python.

 -> Strings are immutable to ensure that they remain constant throughout the program, which leads to better performance and security.
11.	What advantages do dictionaries offer over lists for certain tasks?

 -> Dictionaries allow for fast access and retrieval using keys, which is more efficient for lookups than scanning a list.
12.	Describe a scenario where using a tuple would be preferable over a list.

 -> When you want to store a fixed set of values that shouldn’t change, such as coordinates (x, y), a tuple is preferable.
13.	How do sets handle duplicate values in Python?

 -> Sets automatically remove duplicate entries, storing only unique values.
14.	How does the 'in' keyword work differently for lists and dictionaries?

 -> 'in' checks for element existence in lists and for key existence in dictionaries.
15.	Can you modify the elements of a tuple? Explain why or why not.

 -> No, you cannot modify tuple elements because tuples are immutable.
16.	What is a nested dictionary, and give an example of its use case?

 ->	A nested dictionary is a dictionary within another dictionary. Example: {'student': {'name': 'John', 'age': 21}}.
17.	Describe the time complexity of accessing elements in a dictionary.

 -> Accessing elements in a dictionary has an average time complexity of O(1) due to the hash table implementation.
18.	In what situations are lists preferred over dictionaries?

 ->	Lists are preferred when order matters, or when storing a collection of items without needing key-value mapping.
19.	Why are dictionaries considered unordered, and how does that affect data retrieval?

 ->	Traditionally, dictionaries were unordered collections, so the order of items couldn’t be predicted. However, since Python 3.7, dictionaries maintain insertion order, though conceptually they are still considered unordered.
20.	Explain the difference between a list and a dictionary in terms of data retrieval.

 ->	Lists retrieve data by index (position), while dictionaries retrieve data by key.
